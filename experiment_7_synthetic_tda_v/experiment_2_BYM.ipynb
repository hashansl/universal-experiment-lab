Bayesian Modeling

```python
import pymc as pm
import arviz as az
import pandas as pd
```

```python
n_regions = gdf_county.shape[0]
```

```python
Y = gdf_county['Rate_cou'].to_numpy()

# Expected number of cases per region (constant for simplicity)
E = np.full(n_regions, 30)
```

```python
# Define the model
with pm.Model() as poisson_model_1:
    # Prior for random effect (Normal)
    vN_prior = pm.Normal("vN", mu=0.5, sigma=0.125, shape=n_regions)

    # Assuming you also have a spatial effect uN_prior (add it if it's missing)
    uN_prior = pm.Normal("uN", mu=0, sigma=1, shape=n_regions)  # Adjust parameters as needed

    # Define risk parameter
    thetaN = pm.Deterministic("theta1", pm.math.exp(pm.math.log(E) + uN_prior + vN_prior))

    Y_obs_N = pm.Poisson("Y_obs_N", mu=thetaN, observed=Y)

    # Sample the model

    trace_N = pm.sample(2000, tune=1000, target_accept=0.95, random_seed=42, extend_inferencedata=True)

```

```python
az.plot_trace(trace_N, compact=False);
```

```python
with poisson_model_1:
    pm.compute_log_likelihood(trace_N)
```

```python
pooled_loo = az.loo(trace_N)

pooled_loo
```

```python
# Define the model
with pm.Model() as poisson_model_2:
    # vN is known: add it as a data container
    vN = gdf_county['down_TL_scaled'].to_numpy()
    vN_data = pm.Data("vN", vN)

    # Define the spatial effect as a free parameter
    uN = pm.Normal("uN", mu=0.5, sigma=1, shape=n_regions)

    # Define the risk parameter using the known vN_data and free parameter uN
    thetaN = pm.Deterministic("thetaN", pm.math.exp(pm.math.log(E) + uN + vN_data))
    # thetaN = pm.Deterministic("thetaN", pm.math.exp(pm.math.log(E) + vN_data))


    # Likelihood for the observed data
    Y_obs = pm.Poisson("Y_obs_N", mu=thetaN, observed=Y)

    # Sample from the model
    trace_2 = pm.sample(
        2000, 
        tune=1000, 
        target_accept=0.95, 
        random_seed=42, 
        extend_inferencedata=True
    )
```

```python
with poisson_model_2:
    log_lik = pm.compute_log_likelihood(trace_2, var_names=["Y_obs_N"])
```
