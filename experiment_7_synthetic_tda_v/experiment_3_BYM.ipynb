Bayesian Modeling

```python
import pymc as pm
import arviz as az
import pandas as pd
```

```python
# common variables

n_regions = gdf_county.shape[0]
Y = gdf_county['Rate_cou'].to_numpy()

# Expected number of cases per region (constant for simplicity)
E = np.full(n_regions, 30)
```

```python
# Model 1: Original model

with pm.Model() as poisson_model_original:
    # Prior for random effect (Normal)
    v_prior_original = pm.Normal("v_origina", mu=0.5, sigma=0.125, shape=n_regions)

    # Assuming you also have a spatial effect uN_prior (add it if it's missing)
    u_prior_original = pm.Normal("u_origina", mu=0, sigma=1, shape=n_regions)  # Adjust parameters as needed

    # Define risk parameter
    theta_original = pm.Deterministic("theta_origina", pm.math.exp(pm.math.log(E) + u_prior_original + v_prior_original))

    Y_obs__original = pm.Poisson("Y_obs_original", mu=theta_original, observed=Y)

    # Sample the model

    trace_N = pm.sample(2000, tune=1000, target_accept=0.95, random_seed=42, extend_inferencedata=True)


with poisson_model_original:
    pm.compute_log_likelihood(trace_N)
```

```python
# Model 2: Unstructured effect as down_TL_scaled

with pm.Model() as poisson_model_down_TL_scaled:
    # Prior for random effect (Normal)
    v_down_TL_scaled =  gdf_county['down_TL_scaled'].to_numpy()
    v_down_TL_scaled_data = pm.Data("v_down_TL_scaled", v_down_TL_scaled)

    # Assuming you also have a spatial effect uN_prior (add it if it's missing)
    u_prior_down_TL_scaled = pm.Normal("u_down_TL_scaled", mu=0, sigma=1, shape=n_regions)  # Adjust parameters as needed

    # Define risk parameter
    theta_down_TL_scaled = pm.Deterministic("theta_down_TL_scaled", pm.math.exp(pm.math.log(E) + u_prior_down_TL_scaled + v_down_TL_scaled_data))

    Y_obs_down_TL_scaled = pm.Poisson("Y_obs_down_TL_scaled", mu=theta_down_TL_scaled, observed=Y)

    # Sample the model

    trace_down_TL_scaled = pm.sample(2000, tune=1000, target_accept=0.95, random_seed=42, extend_inferencedata=True)


with poisson_model_down_TL_scaled:
    pm.compute_log_likelihood(trace_down_TL_scaled, var_names=['Y_obs_down_TL_scaled'])
```
