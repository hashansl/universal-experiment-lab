```python
from geodata import generate_grid_dataframes
```

```python
county_df, census_df = generate_grid_dataframes(grid_side_length=7,county_autocorrelation="positive", census_autocorrelation="positive" , random_seed=42)
```

```python
county_df.head(3)
```

```python
census_df.head(3)
```

```python
# plot the county df 

county_df.plot(column='Value', cmap='viridis', legend=True)
```

```python
for i, row in county_df.iterrows():
    county_index = row['Index_county']
    print(f"County {county_index} has value {row['Value']}")

    temp = census_df[census_df['county_id'] == county_index]

    temp.plot(column='Value', cmap='viridis', legend=True)


    # break
```
