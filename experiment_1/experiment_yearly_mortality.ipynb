```python
# load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/h6x/git_projects/universal-experiment-lab/experiment_1/data/total_life_span_mortality_each_year.csv')
```

```python
df.head()
```

```python
# filter by state
df[df['State'] == 'AL']
```

```python
# Convert Variable to a categorical type for ordering
df['Variable'] = pd.Categorical(df['Variable'], ordered=True)

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Variable', y='Total_life_span', marker='o', label='Total Life Span')
sns.lineplot(data=df, x='Variable', y='Sum_Mor', marker='o', label='Sum Morbidity')
plt.title("Trends in Total Life Span and Sum Morbidity")
plt.ylabel("Values")
plt.xlabel("Variable")
plt.legend()
plt.grid()
plt.show()
```

```python
# plot county_count against sum_mor
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='county_count', y='Sum_Mor', hue='State')
plt.title("County Count vs Sum Morbidity")
plt.ylabel("Sum Morbidity")
plt.xlabel("County Count")
plt.show()
```
